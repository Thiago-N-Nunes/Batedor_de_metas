# Batedor de Metas

Ideia inicial: Ter um programa que te diz o quanto é necessario para conquistar uma meta, baseado no tempo em que vc quer bater sua meta

Tecnologias utilizadas: Python



---





Coletando os dados:

```python
meta = input('Digite o nome da sua meta: ')
valor_meta = float(input('Qual o valor que vc quer ter no final da jornada?: R$'))
tempo_meta = int(input('Em quantos meses você quer ter sua meta realizada?: '))
renda = float(input('Qual o valor de suas rendas?:R$ '))
```

 A ideia aqui é conseguir imprimir 2 cenários, possibilitando que o usuário tenha uma saude financeira e mesmo assim conquistar sua meta. Serão passados 2 cenários: Guardando 10% da renda e guardando com base no tempo em que o usuário deseja

Cenário 1

```python
def meta_saudavel():
  qtd_saudavel = renda / 100 * 10
  meses_meta = int(valor_meta / qtd_saudavel)
  valor_final = qtd_saudavel * meses_meta
  print('CENÁRIO 1')
  if qtd_saudavel < valor_meta:
    valor_restante = valor_meta - valor_final
    print(f'Levando em consideração sua saúde financeira, separamos um cenário em que você guarde 10% de sua renda afim de ter sua meta batida, nesse cenário, você guardaria R${qtd_saudavel} por {meses_meta} meses, e mais R${valor_restante:.2f} no ultimo mês')
  else:
    print(f'Levando em consideração sua saúde financeira, separamos um cenário em que você guarde 10% de sua renda afim de ter sua meta batida, nesse cenário, você guardaria R${qtd_saudavel} por {int(meses_meta)} meses')
```

Cenário 2

```python
def meta_esperada():
  qtd_desejada = valor_meta / tempo_meta
  print('CENÁRIO 2')
  print(f'Já que sua meta é juntar R${valor_meta} em {tempo_meta} meses, você precisa guardar R${qtd_desejada:.2f} por mês')
```

Resultado do projeto

```python
meta_saudavel()
meta_esperada()
```
